```python
# Define the training data
import numpy as np

# X_train = np.random.random((5000, 32))
# y_train = np.random.random((5000,1))

mean = 0.0
std_dev = 1.0

# Generate an array of random numbers from a normal distribution
X_train = np.random.normal(mean, std_dev, size=(5000, 32))
y_train = np.random.normal(mean, std_dev, size=(5000, 1))

```

```python
# Define the neural network model
# sequential/functional
from keras import models
# dense, dropout
from keras import layers

INPUT_DIM = X_train.shape[1]

model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_dim=INPUT_DIM))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(4, activation='sigmoid'))
model.add(layers.Dense(1, activation='softmax'))
```

```python
model.summary()
```

```python
# Configure the learning process
from keras import optimizers
from keras import metrics

model.compile(optimizer='adam',
              loss='mse',
              metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train,
          batch_size=100,
          epochs=100)
```
